```python
from datetime import datetime
```

```python
# Строки с датами
moscow_times = "Wednesday, October 2, 2002"
the_guardian = "Friday, 11.10.13"
daily_news = "Thursday, 18 August 1977"  # исправлено: 19771 → 1977

# Форматирование и преобразование
dt_moscow = datetime.strptime(moscow_times, "%A, %B %d, %Y")
dt_guardian = datetime.strptime(the_guardian, "%A, %d.%m.%y")
dt_daily = datetime.strptime(daily_news, "%A, %d %B %Y")

# Вывод результатов
print(dt_moscow)
print(dt_guardian)
print(dt_daily)
```
