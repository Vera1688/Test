```python
import clickhouse_connect
```

```python
client = clickhouse_connect.get_client(
    host='ваш_хост',
    port=8443,
    username='ваш_логин',
    password='ваш_пароль',
    secure=True,
    verify=True
)
```

```python
# SQL‑запрос
query = """
SELECT
    m.title AS movie_title,
    COUNT(r.rating) AS count_5_ratings
FROM ratings AS r
JOIN movies AS m ON r.movieId = m.movieId
WHERE r.rating = 5.0
GROUP BY m.title
ORDER BY count_5_ratings DESC
LIMIT 1
"""
```

```python
# Выполняем запрос
result = client.query(query)
rows = result.result_rows
```

```python
# Выводим результат
if rows:
    movie_title, count = rows[0]
    print(f"Фильм: {movie_title}")
    print(f"Количество оценок 5.0: {count}")
else:
    print("Данных не найдено.")

# Закрываем соединение
client.close()
```
