```python

import requests
import json
```

```python

# URL сервиса ЦБ РФ с курсами валют в JSON
url = "https://www.cbr-xml-daily.ru/daily_json.js"
```

```python
# Получаем данные
response = requests.get(url)
data = response.json()

currencies = data['Valute']

max_currency = None
max_value = 0

for code, currency in currencies.items():
    if currency['Value'] > max_value:
        max_value = currency['Value']
        max_currency = currency['Name']
print(f"Валюта с максимальным курсом: {max_currency}")
print(f"Курс: {max_value} руб.")
```
