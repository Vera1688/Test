```python
import re

def validate_car_id(car_id):
    valid_letters = 'АВЕКМНОРСТУХ'
  
    pattern = rf'^([{valid_letters}])(\d{{3}})([{valid_letters}]{{2}})(\d{{2,3}})$'
    
    
    match = re.match(pattern, car_id.strip())
    
    
    if match:
        letter1 = match.group(1)
        digits3 = match.group(2)
        letters2 = match.group(3)
        region = match.group(4)
        
        number = f"{letter1}{digits3}{letters2}"
        return f"Номер {number} валиден. Регион: {region}."
    else:
        return "Номер не валиден."

# Примеры использования
print(validate_car_id('А222ВС96'))   # Номер А222ВС валиден. Регион: 96.
print(validate_car_id('АБ22ВВ193')) 
```
